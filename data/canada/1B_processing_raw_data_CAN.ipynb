# Choosing Data
In this notebook, we choose a subset of the "COVID-19 Twitter Dataset with Latent Topics, Sentiments and Emotions Attributes" dataset for Canada[1]. We do this since we have a limited amount of compute at our disposal. The full dataset contain tens of millions of rows.


```python
import pandas as pd
```

```python
# Import CSV data file
data = pd.read_csv("files/project_tweet_data_CAN.csv")

# Check first five rows
data.head(5)
```

```python
# Check last five rows
data.tail(5)
```

```python
# Check shape
data.shape
```

```python
# Take initial subset
data_subset = data.head(2252806)
```

```python
# Reset indicies
data_subset = data_subset.reset_index(drop=True)
```

```python
# Inspect last five rows
data_subset.tail(5)
```

```python
### We still have too many rows.. ##

# Take every 3th row and discard the rest
data_new = data_subset.iloc[::3, :]

# Reset indicies
data_new = data_new.reset_index(drop=True)

# Inspect shape of new subset DataFrame
data_new.shape
```

```python
# Save to CSV
data_new.to_csv("files/project_tweet_data_final_CAN.csv")
```

```python
# Import saved CSV file to inspect that it is in order
data_import = pd.read_csv("files/project_tweet_data_final_CAN.csv",index_col=0)
```

```python
# Inspect first five elements
data_import.head(5)
```

```python
# Inspect final five rows
data_import.tail(5)

## Looks OK! ##
```
