```python
import pandas as pd
import networkx as nx
import numpy as np
import re
```

```python
uk_df = pd.read_csv("files/cleaned_twitter_data_CAN.csv", index_col = 0, dtype={'user_id': str})
```

```python
uk_df
```

```python
def find_mention(tweet):
    pattern = r'@([A-Za-z0-9_]+)'
    match = re.findall(pattern, tweet)

    return match
```

```python
uk_df = uk_df.dropna(subset = ["tweet"])
```

```python
uk_df
```

```python
uk_df = uk_df.reset_index(drop = True)
```

```python
np.sum(uk_df['tweet'].isnull().values)
```

```python
uk_df['mentions'] = uk_df['tweet'].apply(find_mention)
```

```python
uk_df['mentions']
```

```python
mentions_df = uk_df[uk_df['mentions'].map(lambda d: len(d)) > 0]
```

```python
mentions_df
```

```python
mentions_df.to_csv('files/mentions_CAN.csv')
```
